```python
from transformers import GPT2Tokenizer

tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

text = "Sitting here lost in my thoughts, it suddenly occurred to me that"

tokens = tokenizer.tokenize(text)
token_ids = tokenizer.encode(text)

print("Tokens:", tokens)
print("Token IDs:", token_ids)
print("Number of tokens:", len(tokens))
```

```python
import pandas as pd

# Change this to your CSV filename
csv_path = "the_almost_ultimate_final(trimming_concepts).csv"

# Read CSV
df = pd.read_csv(csv_path)

# Make sure the required column exists
if "Concept" not in df.columns:
    raise ValueError("CSV must contain a column named 'Concept'.")

# Read the column as strings, dropping missing values
concepts = df["Concept"].dropna().astype(str).tolist()

# Tokenize each concept separately and store lengths
results = []
for concept in concepts:
    token_ids = tokenizer.encode(concept)
    results.append((concept, len(token_ids)))

# Sort by token length descending
results.sort(key=lambda x: x[1], reverse=True)

# Print the top 10 concepts with maximal tokenization length
top_10 = results[:10]

print("Top 10 concepts by GPT-2 token length:\n")
for i, (concept, length) in enumerate(top_10, 1):
    print(f"{i}. [{length} tokens] {concept}")

# Print the maximal length
if results:
    print("\nMaximal tokenization length:", results[0][1])
else:
    print("\nNo concepts found.")
```

```python
def print_token_lengths(sentences, tokenizer):
    results = []

    for s in sentences:
        token_ids = tokenizer.encode(s)
        results.append((s, len(token_ids)))

    # Sort by length (descending)
    results.sort(key=lambda x: x[1], reverse=True)

    print("Sentences sorted by token length (longest → shortest):\n")
    for s, length in results:
        print(f"[{length}] {s}")


# Example usage:
print_token_lengths(sentences, tokenizer)
```
